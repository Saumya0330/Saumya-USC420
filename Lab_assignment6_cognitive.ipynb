Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

M=int(input("Enter M value: "))
x=np.linspace(-10,10,100)
y1=M*(x**2)
y2=M* np.sin(x)

#print(x)
plt.plot(x,y1, 'r--', label="y=M*x^2")
plt.plot(x,y2, 'b-', label="y=M*sin(x)")
plt.legend()
plt.grid()
plt.title('Mathematical functions')
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import seaborn as sns

data = {
    'Subject': ['Math', 'Physics', 'Chemistry', 'Biology', 'English'],
    'Score': [85, 92, 78, 88, 95]
}
df=pd.DataFrame(data)
sns.set_style('whitegrid')
ax=sns.barplot(x='Subject',y='Score',data=df, legend=False , hue='Subject', palette='pastel')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}',(p.get_x()+p.get_width()/2, p.get_height()), ha='center', va='bottom')

plt.title('Subject Scores')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid()
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
roll_no=int(102317028)
np.random.seed(roll_no)
values=np.random.randn(50)
fig, ax =plt.subplots(2,2)

ax[0,0].plot(range(50),np.cumsum(values), marker='o', alpha=0.5)
ax[0,0].set_title('Cumulative Sum Line Plot')
ax[0,0].set_xlabel('Index')
ax[0,0].set_ylabel('Ccumulative Sum')
ax[0,0].grid()

ax[0,1].scatter(range(50),values, alpha=0.6 )
ax[0,1].set_title('Scatter Plot with Random Noise')
ax[0,1].set_xlabel('Index')
ax[0,1].set_ylabel('Random Value')
ax[0,1].grid()

ax[1,0].axis('off')
ax[1,1].axis('off')

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
df=pd.read_csv('company_sales_data.csv')
#print(df.head)

sns.lineplot(x='month_number', y='total_profit', data=df)
plt.title('Total Profit over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid()
plt.show()

product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in product_columns:
    sns.lineplot(data=df, x='month_number', y=product, marker='o', label=product)
plt.title('Product Sales Data Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend(title='Products')
plt.grid()
plt.show()

total_sales = df[product_columns].sum()
sns.barplot(x=total_sales.index, y=total_sales.values, palette='viridis', hue=total_sales.index)
plt.title('Total Sales of Each Product')
plt.xlabel('Products')
plt.ylabel('Total Sales Units')
plt.grid(axis='y')
plt.show()
```
